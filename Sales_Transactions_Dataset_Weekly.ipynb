# Elbow Curve 

### Find Optimal Number of Clusters

```python
from sklearn.cluster import KMeans
```

```python
wcss= []

for k in range(1,11):
    kmeans = KMeans(n_clusters = k , init='k-means++')
    kmeans.fit(x)
    wcss.append(kmeans.inertia_)
print(wcss)
```

```python
plt.figure(figsize=(12,14))
plt.plot(range(1,11),wcss)
plt.xticks(range(1,11))
plt.show()
```

By See Elbow Curve Method We can Identify the value of K = 4

```python
from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters=4,init='k-means++')
kmeans.fit(x)
```

```python
y_meansPred = kmeans.predict(x)
```

```python
y_meansPred
```

```python
df['cluster '] = pd.DataFrame(y_meansPred)
df
```

# Hierarchical Clustering 

```python
import scipy.cluster.hierarchy as sch

df1 = sch.linkage(x,method = 'ward')

dendrogram = sch.dendrogram(df1)

plt.title('Dendrogram')
plt.xlabel('customers')
plt.ylabel('Euclidean distances')
plt.show()
```

```python
from sklearn.cluster import AgglomerativeClustering 

hc = AgglomerativeClustering(n_clusters = 5, affinity = 'euclidean',linkage = 'ward')

y_hc = hc.fit_predict(x)
```

```python
y_hc
```

```python
df['Hierarchical_Clustering'] = pd.DataFrame(y_hc)
df
```

# DBSCAN Clustering 

```python
from sklearn.cluster import DBSCAN

dbs = DBSCAN(eps=2,min_samples=3)

y_dbs = dbs.fit_predict(x)
```

```python
df['DBSCAN_Clustering'] = pd.DataFrame(y_dbs)
```
